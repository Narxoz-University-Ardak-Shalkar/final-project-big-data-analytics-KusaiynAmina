```python
import pandas as pd

# Загрузка данных
data = pd.read_csv("test.csv")

# Обработка пропущенных значений
data['Age'] = data['Age'].fillna(data['Age'].median())  # Заполняем пропуски медианой
data['Fare'] = data['Fare'].fillna(data['Fare'].mean())  # Заполняем пропуски средним значением

# Удаление столбца с большим количеством пропусков
if 'Cabin' in data.columns:  # Проверяем, существует ли столбец перед удалением
    data = data.drop(columns=['Cabin'])

# Проверка на дубликаты
data = data.drop_duplicates()

# Преобразование категориальных данных в числовые
if 'Sex' in data.columns:
    data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

# Вывод первых нескольких строк для проверки
print(data.head())

```

```python
# Статистика числовых данных
print(data.describe())

# Частотный анализ категориальных переменных
print(data['Sex'].value_counts())
print(data['Embarked'].value_counts())


# Выбираем только числовые столбцы для расчета корреляции
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Рассчитываем корреляционную матрицу
corr = numerical_data.corr()

# Визуализация корреляционной матрицы
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()

```

```python
# Создание новых признаков
data['FamilySize'] = data['SibSp'] + data['Parch']

# Бинаризация возраста (например, "молодой" и "взрослый")
data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 18, 60, 100], labels=['Child', 'Adult', 'Senior'])

# Анализ влияния класса и возраста на стоимость билета
pivot_table = data.pivot_table(index='Pclass', columns='AgeGroup', values='Fare', aggfunc='mean')
print(pivot_table)
```

```python
# График распределения по полу
data['Sex'].value_counts().plot(kind='bar')
plt.title('Распределение по полу')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.show()

# График распределения по классу
data['Pclass'].value_counts().plot(kind='bar')
plt.title('Распределение по классу')
plt.xlabel('Класс')
plt.ylabel('Количество')
plt.show()

# Визуализация FamilySize
sns.boxplot(data=data, x='FamilySize', y='Fare')
plt.title('Зависимость стоимости билета от размера семьи')
plt.show()
```

```python
# Использование apply() для создания нового признака
def get_title(name):
    if "Mr." in name:
        return "Mr"
    elif "Mrs." in name:
        return "Mrs"
    elif "Miss" in name:
        return "Miss"
    else:
        return "Other"

data['Title'] = data['Name'].apply(get_title)

# Сводная таблица по новым признакам
pivot = data.pivot_table(index='Title', columns='Pclass', values='Fare', aggfunc='mean')
print(pivot)
```
